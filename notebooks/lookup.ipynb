```python
import pandas as pd
df = pd.read_parquet('../data/processed/storm_featured.parquet')

# What damage_tier do Florida Tornado rows actually have?
mask = (df['STATE'] == 13) & (df['EVENT_TYPE'] == 46)
print(df[mask]['damage_tier'].value_counts())
print(f"Total Florida Tornado rows: {mask.sum()}")
```

```python
mask = (df['STATE'] == 33) & (df['EVENT_TYPE'] == 24)
print(df[mask]['damage_tier'].value_counts())
print(f"Total: {mask.sum()}")
```
